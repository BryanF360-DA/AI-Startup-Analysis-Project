```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://builtin.com/artificial-intelligence/ai-companies-roundup'

print("Attempting to fetch the webpage...")

try:
    response = requests.get(url)
    response.raise_for_status() # This checks for any errors in the request
    print("Success! Webpage content fetched.")
except requests.exceptions.RequestException as e:
    print(f"Failed to fetch webpage. Error: {e}")
    # Parse the HTML content into a BeautifulSoup object
soup = BeautifulSoup(response.content, 'html.parser')

print("HTML successfully parsed. The 'soup' object is now ready.")
# Find all the h3 tags, which contain the company names
company_tags = soup.find_all('h3')

# Print the number of tags found to verify
print(f"Found {len(company_tags)} h3 tags.")
# Create an empty list to store the clean names
company_names = []

# Loop through each tag in the company_tags list
for tag in company_tags:
    # Get the text from the tag and add it to our new list
    company_names.append(tag.get_text())

# Print the first 10 items to check our work
print("First 10 extracted names:")
print(company_names[:10])
# Clean the list by removing the last item (the unwanted header)
cleaned_company_names = company_names[:-1]

# Create the final DataFrame from our clean list
df = pd.DataFrame(cleaned_company_names, columns=['CompanyName'])

# --- FINAL VALIDATION ---
# Print a success message and show the first 5 rows of our new table
print(f"Module 2 Complete: Successfully created a DataFrame with {len(df)} companies.")
print("DataFrame Head:")
display(df.head())
```
